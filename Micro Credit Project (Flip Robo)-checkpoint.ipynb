```python
#removing skewness
micro.skew()
```

```python
#feature selection
x=micro.drop(columns=['label'],axis=1)
y=micro['label']
```

```python
#removing skewness
from sklearn.preprocessing import power_transform
```

```python
x=power_transform(x,method='yeo-johnson')
print(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.5,random_state=42,stratify=y)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
gd=GradientBoostingClassifier()
gd.fit(x_train,y_train)
pregd=gd.predict(x_test)
print(confusion_matrix(y_test,pregd))
print(classification_report(y_test,pregd))
print(accuracy_score(y_test,pregd))
```

```python
rd=RandomForestClassifier()
rd.fit(x_train,y_train)
prerd=rd.predict(x_test)
print(confusion_matrix(y_test,prerd))
print(classification_report(y_test,prerd))
print(accuracy_score(y_test,prerd))
```

```python
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
predtc=dtc.predict(x_test)
print(confusion_matrix(y_test,predtc))
print(classification_report(y_test,predtc))
print(accuracy_score(y_test,predtc))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cvsscore=cross_val_score(RandomForestClassifier(),x,y,cv=5,scoring='accuracy')
print(cvsscore)
print(cvsscore.mean())
print(cvsscore.std())
```

Observation-RandomForestClassifier is working best amongst all.
